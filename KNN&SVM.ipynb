```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python

# Load dataset
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
# Mapping target values to species names
species_map = {0: "setosa", 1: "versicolor", 2: "virginica"}
df['target'] = [species_map[i] for i in iris.target]

  # Target labels

# Display first 5 rows
df.head(104)

```

```python
X =df.iloc[:, [2, 3]].to_numpy()
X
```

```python
Y =df.iloc[:,4].to_numpy()
Y
```

```python
##split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=7)
```

```python
print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
#fit
clf = KNeighborsClassifier()
clf.fit(X_train,Y_train)
```

```python
score = clf.score(X_train,Y_train)
score
```

```python

print(clf.predict([[5.6,2.4]]))
```

```python
predict = clf.predict(X_test)
print('oreginal',Y_test)
print('Predict',predict)
```

```python
print(classification_report(Y_test,predict)) #precision compare= predict to oreginal, recall compare=oreginal to predict   , support = number of o , 1 ,2
```
